```python
def files_to_list (directory = r"D:\Data_scripts\MUA_doubleshank_Neuronexus_probes\2020-12-07", filename_ending = "Rec_7_201207_132859.rhs" ):
    """returns absolute file_paths of data files as list of strings
    
    Parameters
    ----------
    directory: specify the directory where files are located; must be string or converted to raw string (r'...')
    
    filename_ending:  specifiy the file ending e.g. '.npy' , 'NOT_rm.npy'
    """
    
    ####change to directory where files are located, if system can not find the path try to convert to raw string, r'....'
    os.chdir(directory)
    ####empty list for filepaths: file_path
    file_path = []
    file_name = []
    ####for loop appends filename with specific ending to file_path
    for filename in os.listdir(directory):
       if filename.endswith(filename_ending):
           ###creates absolute filepath by merging directory with filename
           file_path.append(os.path.join(directory, filename))
           file_name.append(filename)
    return file_path, file_name

def create_directory (parent_dir = '_', directory= '_'):
    """create a directory within a specified parent directory"""
    path = os.path.join(parent_dir, directory)
    os.mkdir(path)


def load_digin (filepath, sampling_rate = 20000):
    """ returns digin_list--> digital input timestamps in samplepoints (from e.g. PCI stim) for slicing data into single trials. 
    
    Parameters:
    ----------    
    Copy absolute filepath from file explorer and pass as string: "filepath" """
        ############get digital inputs for PCI stimulation###########
    #global digin_list, file_path, digin_half2, digin_half2_samples
    file = pyintan.File(filepath)
    digin = file._digital_in[0].times
    ######digin contains 2 timestamps for every trigger signal; previously these 2 timestamps were in consecutive order. Now 1. timestamp for every trigger is in 1. half of array,  2. timestamp for every trigger is in 2. half of array  ####
    #find start index of 2nd half of digin array: ix
    ix = int(len(digin)/2)
    digin_half2= digin[ix:]
    file_path = filepath
    del file
    ######convert digin_half2 (s) into corresponding sample points: store in digin_half2_samples #############
   
    digin_half2_samples= digin_half2*sampling_rate
    #######pyintan provides digital_in[0].times as array as type "quantity"; -->can't be used in for loop for list slicing--> import quantities and convert with.item() function#####
    digin_list = [digin_half2_samples[i].item() for i, j in enumerate(digin_half2_samples)]  
    ##########convert digin_list float valuesto int--> necessary for indexing in for loop#######
    digin_list=[round(x) for x in digin_list]
    
    return digin_list, file_path, digin_half2, digin_half2_samples
  
   
      
def slice_array_for_conv (data_lfp, sampling_rate = 800, interval = 1600, retain_digin = 1):    
    """returns data: n-dimensional array (samples x channels x trials), where data are centered around the diginput (specified by interval in samples)
    
    Parameters
    ----------
    data: input array
    
    sampling_rate: provide the sampling rate of the raw or previously filtered (and downsampled) data
    
    interval: provide sample interval (int) around stimulation time (according to digin_list). If none are given the default value is 1600 samples 
    
    retain_digin: default=1,  retains every nth digin from the digin list
    
    """
    global digin_list_lfp
    
    digin_half2_samples= round(digin_half2 * sampling_rate).astype(int)
    #######pyintan provides digital_in[0].times as array as type "quantity"; -->can't be used in for loop for list slicing--> import quantities and convert with.item() function#####
    digin_list_lfp = [digin_half2_samples[i].item() for i, j in enumerate(digin_half2_samples)]  
    
    ###retain very nth digin entry retain_digin
    digin_list_lfp = digin_list_lfp[0::retain_digin]
    
    ########create data matrix  with samples x trials x channels: data
    data = np.zeros((interval*2+1,np.size(digin_list_lfp), np.size(data_lfp, axis =0)), dtype =np.float16)

   
    #for ch in range (np.size(data, axis =1)): 
    for i in range (np.size(data_lfp, axis=0)):
        for ii in range (np.size(digin_list_lfp, axis =0)):
            x = digin_list_lfp[ii] - interval
            y = digin_list_lfp[ii] + interval + 1 #+1 includes an extra sample point --> useful when defining time vectors with start and stop values included (e.g. range from -2 to 2, consists of 5 samples then)
            
            data[:,ii,i] =  data_lfp[i][x:y]
    #return array as samples x channels x trials
    data = np.transpose(data, (0,2,1))
    return data


def channel_rejection(data, sampling_rate = 20000, poststim_rms_x1 = 0.1, poststim_rms_x2 = 0.5, dependent_rms_x1 = 0.1, dependent_rms_x2 = 0.5, factor_std =3):
    """returns data (samples x channels x trials) where channels with high threshold voltage artefacts are removed. 
   
    calculates:
    1) the averaged root mean square (RMS) after stimulation for each channel 
    2) the median over all channels and added to the standard deviation of the median multiplied by factor _std: threshold for channel rejection
   
    
    Parameters
    ----------
    data: input array, numpy array (samples x channels x trials)
    
    poststim_rms_x1: in s, start for time range of baseline used to compute the rms for each trial rel. to stim artefact
    
    poststim_rms_x2: in s, stop for time range of baseline used to compute the rms for each trial rel. to stim artefact    
        
    dependent_rms_x1: in s, start time range of baseline used to compare [s]
    
    dependent_rms_x2: in s, stop time range of baseline used to compare [s]
    
    factor_std: (int), scale factor multiplies with std of rms to calculate threshold for channel rejection. -->high factor rejects less channels
    """
    global median_ch, median_std, threshold, mean_rms_post, mean_rms_post_dependent, prestim, start1, stop1, start2, stop2, poststim_matrix, rms_post, std_ch, median_ch
    
    
    #number of samples from start to stimulation artefact: prestim
    prestim = (np.size(data, axis =0)/2)/sampling_rate
    
    #start and stop samples of baseline (after stimulation) for calculating rms: start1, stop1
    start1 = int(sampling_rate * poststim_rms_x1)+ int(sampling_rate * prestim)
    stop1  = int(sampling_rate * poststim_rms_x2)+ int(sampling_rate * prestim)
        
    #start and stop samples of baseline (after stimulation) for comparison with rms: start2, stop2
    start2 = int(sampling_rate * dependent_rms_x1)+ int(sampling_rate * prestim)
    stop2  = int(sampling_rate * dependent_rms_x2)+ int(sampling_rate * prestim)

    
    
    ## select the range to calculate the root mean square of baseline: poststim_matrix
    poststim_matrix = data[start1:stop1,:,:].astype('float64')  
          
    ## select the range for comparison with rms: dependent_matrix
    dependent_matrix = data[start2:stop2,:,:].astype('float64')    

    
    ##calculate root-mean-squared of prestim_matrix, dependent_matrix : rms_post, rms_post_dependent (array shape: channels x trials)
    rms_post           = np.sqrt(np.mean(poststim_matrix**2, axis = 0))
    
    rms_post_dependent = np.sqrt(np.mean(dependent_matrix**2, axis = 0))             
    
    
    ## DEFINES THRESHOLD FOR REJECTION
        
    ## Mean and of RMS from all baselines from all trials : mean_rms_post, mean_rms_post_dependent (array shape: channels)
    mean_rms_post = np.mean(rms_post, axis =1)
    
    mean_rms_post_dependent = np.mean(rms_post_dependent, axis = 1)
   
        
    ##calculate median of all channels: median_ch (scalar)
    median_ch = np.median(mean_rms_post)
    
    ##calculate std of rms_post over all trials: std_ch (array: channels)
    std_ch = np.std(rms_post, axis = 1)
    
    ##calculate median std of std_ch: median_std (array: channels)
    median_std = np.median(std_ch) 
    
        
    # threshold at 3 standard deviations : threshold 
    threshold = median_ch + (median_std * factor_std)
    
    
    
   ##### PERFORMS REJECTIONS AND CONSERVES channels 
   ###Identifies only those channels having baseline RMS below threshold 
   
    #create empty array of shape mean_rms_post for appending channels: a1
    a1 = np.zeros(np.shape(mean_rms_post))
   
      
   #loop over channels and trials:
    for ch in range (np.size(rms_post_dependent, axis = 0)):
        
        #rms values samller than threshold are appended as True to a1 array (larger values as False)
        a1[ch] = mean_rms_post_dependent[ch]    <   threshold 
        
        #find chanel indices where a1==True: channels_find
        channels_find = np.where(a1 ==True)[0]
    
        
    
   #Extract only channels with baseline RMS below threshold  
    eeg_good = data[:,channels_find, :]
    return eeg_good
                     
    

def rereferencing (data, return_medref = False, return_avref = False):
    """returns the rerefenced data (eeg_avref, eeg_medref)  
    by subtracting the common mean, median of all channels
    
    Parameters
    ----------
    data: input data (channels x samples) returned by recording_prb.get_traces()  
    
    return_medref: if True, the common median referenced data are returned
    
    return_avref: if True, the common average referenced data are returned
    """
    
    ###re-reference the data 
    
    if return_avref  == True:
        ###computes the common average across all channels: average_ch
        average_ch = np.squeeze(np.mean(data, axis = 0))
        ###create matrix for mean referenced data: eeg_avref
        eeg_avref =  np.zeros((np.shape(data)))
        
            ###loop over all channels to perform subraction
        for i in range(np.size(data, axis =0)):
            
            ###common average substraction
            eeg_avref[i,:] = data[i,:] - average_ch

    
    if return_medref == True:
        ###computes the common median across all channels: median_ch
        median_ch = np.squeeze(np.median(data, axis =0))
        ###create matrix for median referenced data: eeg_medref
        eeg_medref = np.zeros((np.shape(data)))   
    
        ###loop over all channels to perform subraction
        for i in range(np.size(data, axis =0)):
            
            ###common median subtraction
            eeg_medref[i,:] = data[i,:] - median_ch
            
    if return_avref  == True:    
    
        return eeg_avref
    
    if return_medref  == True:    
    
        return eeg_medref


    
def interpolate_rm_artifact_linear(data, sampling_rate=20000, ms_before=3, ms_after=5):
    """returns data: n-dimensional array, where stimulation artifact was removed using interpolation
    Parameters
    ----------
    data: input array (returned from 'slice_array_for_conv' function)
    
        
    ms_before= 2: artifact will be interpolated from  x ms before.  -->(to check how many samples_before and samples_after should be chosen, plot data and determine  visually)
    
    ms_after= 4: artifact will be interpolated from  x ms after.
   
   
    
    """
    
    samples_before = int(sampling_rate * ms_before /1000)
    samples_after = int(sampling_rate * ms_after /1000)
    ###the stimulation artefact is in the middle of the data array. get artefact sample point by dividing data axis by two: artefact
    artefact = int(np.size(data, axis =0)/2)
    for i in range (np.size(data, axis =1)):
       
        x = artefact - samples_before
        y = artefact + samples_after
        
        yinterp = np.squeeze(np.interp([range(x, y)], [x, y], [data[i, x], data[i, y]]))
        
        ###add noise to yinterp, based on random samples 5 ms (0.005s) before x: yinterp_noise
        ##define baseline to choose random samples from : pre_x
        pre_x = int(0.002 * sampling_rate)
        
        #yinterp_noise = np.random.choice(data[(x-pre_x):x,i,ii], size = len(yinterp))/3
        
        
        ###add yinterp_noise to yinterp and insert where artefact was before
        data[i, x:y] = yinterp 
    
    return data


def interpolate_rm_artifact_spline(data, sampling_rate=20000, x_ms=3, y_ms=5, x1_ms=10, y1_ms =3, smooth = 0, der= 0, k =3):
    """returns data: n-dimensional array, where stimulation artifact was removed using interpolation
    Parameters
    ----------
    data: input array (returned from 'slice_array_for_conv' function)
    
        
    x_ms: artifact will be interpolated from  x ms before.  -->(to check how many samples_before and samples_after should be chosen, plot data and determine  visually)
    
    y_ms: artifact will be interpolated from  x ms after.
    
    x1_ms: the baseline used for calculating the interpolation before the artefact
    
    y1_ms: the baseline used for calculating the interpolation after the artefact
     
    Background: REMOVAL OF STIMULUS ARTEFACT
    The intracranial electrical stimulation normally introduces 
    an artefactual fast and high voltage transient that needs to be removed from 
    electrophysiological recording before filtering. 
    The exact samples corresponding to the onset of all the stimulations are identified and 0.005 s 
    of the signal from those time points are removed and interpolated with a spline function in all channels.
    
    """
    
    #global x_for_interp
    ###Parameters
    #data = recording_trace
    #sampling_rate=20000
    #x_ms=3
    #y_ms=5
    #x1_ms=10
    #y1_ms =3
    #smooth = 0
    #der= 0
    #k =3
    
    
    ###the timestamps of the stimulation artefacts correspond to entries in digin_half2_samples : artefact
    artefact = digin_list

    for artefacts in artefact:    
    
        ###the range for interpolation: x_samples, y_samples
        x_samples = int(sampling_rate * x_ms /1000)
        y_samples = int(sampling_rate * y_ms /1000)
        
        ###the data range before and after interpolation: x1_samples, y1_samples
        x1_samples = int(sampling_rate * x1_ms /1000)
        y1_samples = int(sampling_rate * y1_ms /1000)
        
        ###the range for interpolation: x2_samples, y2_samples
        x2_samples = artefacts - x_samples
        y2_samples = artefacts + y_samples
                    
        ###range for interpolation baseline : x3, y3
        x3_samples = x2_samples - x1_samples
        y3_samples = y2_samples + y1_samples
        
        ###x_data vector for interp:x_data
        x_data = np.arange(np.size(data, axis =1))
        
            
        ###subset of x_data for interpolation, baseline before artefact: x3_x2_data
        x3_x2_data = x_data[x3_samples:x2_samples]
        
        ###subset of x_data for interpolation, baseline after artefact: y3_x2_data
        y3_y2_data = x_data[y2_samples:y3_samples]
        
         
        
    
        ###append x data before and after artefact to array: x_for_interp
        x_for_interp = np.append(x3_x2_data, y3_y2_data)
        
        
        
        for i in range (np.size(data, axis =0)):            
                    
            ###append y data before and after artefact to array: y_for_interp
            y_for_interp = np.append(data[i, x3_samples:x2_samples], data[i, y2_samples:y3_samples])
    
            tck = interpolate.splrep(x_for_interp, y_for_interp, s = smooth, k = k)
            
            x_new = x_data[x3_samples : y3_samples]
            
            yinterp = interpolate.splev(x_new, tck, der = der)
            ###yinterp without baseline: yinterp_new
            yinterp_new = yinterp[x1_samples:-y1_samples]
            
            data[i, x2_samples : y2_samples] = yinterp_new
                   
    return data


def notch_filter (data, freq = 50, sampling_rate = None,  qfactor= 3):
    """returns notch filtered data (samples x channels x trials) using scipy iirnotch notch filter) and filtfilt function. 
    
    Parameters
    ----------
    data: input data as numpy array (samples x channels x trials)
    
    sampling_rate: sampling rate of data
    
    freq: the frequency  of the notch filter to remove ( 50 Hz)
     
    qfactor: Quality factor. Dimensionless parameter that characterizes notch filter -3 dB bandwidth bw relative to its center frequency, Q = w0/bw.


    
    """
    ###sets butteworth filter
    
     ###the nyquist frequency for the given sample rat: nyq
    nyq = sampling_rate/2   
    
    ###notch frequency passed as normalized scalar: f_notch
    f_notch = freq/nyq
     
     
        
    #set highpass filter: b,a
    b,a = sp.signal.iirnotch(freq, qfactor, sampling_rate)
     
   
     
    #create empty arrays to append filtered data
    eeg_a = np.zeros(np.shape(data))
     
     ###performs filtering using filtfilt which filters in forward and backward directions to avoid phasedelay introduced by filtering in only one direction
    for ch in range(np.size(data, axis = 1)):
        for sw in range(np.size(data, axis = 2)):
            eeg_a[:,ch,sw] = sp.signal.filtfilt(b,a,data[:,ch,sw])
             
    return eeg_a
    


def bandpass_filter (data, sampling_rate = 20000, freq_low= 80, freq_high = 0.5, order = 3):
    """returns bandpass filtered data (samples x channels x trials) using scipy butter (Butterworth) and filtfilt function. 
    
    Parameters
    ----------
    data: input data as numpy array (samples x channels x trials)
    
    sampling_rate: sampling rate of data
    
    freq_low: the low pass frequency of the filter (the upper frequency band, e.g. 80Hz)
        
    freq_high: the high pass frequency of the filter (the lower frequency band, e.g. 0.5 Hz)
    
    order: the order of the filter
    
    """
    ###sets butteworth filter
    
     ###the nyquist frequency for the given sample rat: nyq
    nyq = sampling_rate/2    
       
     
     
    #low pass frequency for butterworth: wn_low
    wn_low  = freq_low/nyq     
        
     
    #high pass frequency for butterworth: wn_high
    wn_high  = freq_high/nyq
     
    #set highpass filter: b,a
    b,a = sp.signal.butter(order, freq_low, btype = 'lowpass', fs=sampling_rate)
     
    #set lowpass filter: d,c
    d,c = sp.signal.butter(order, freq_high, btype = 'highpass', fs=sampling_rate)   
     
    #create empty arrays to append filtered data
    eeg_a = np.zeros(np.shape(data))
    eeg_b = np.zeros(np.shape(data))
     
     ###performs filtering using filtfilt which filters in forward and backward directions to avoid phasedelay introduced by filtering in only one direction
    for ch in range(np.size(data, axis = 1)):
        for sw in range(np.size(data, axis = 2)):
            eeg_a[:,ch,sw] = sp.signal.filtfilt(b,a,data[:,ch,sw])
            eeg_b[:,ch,sw] = sp.signal.filtfilt(d,c,eeg_a[:,ch,sw])
             
    return eeg_b
    
 
def downsampling (data, sampling_rate = 20000, sampling_rate_new = 500):
    """return downsampled data (samples x channels x trials)
    
    Parameters
    ----------
    data: input data, numpy array (samples x channels x trials)
    
    sampling_rate: the sampling rate of input data
    
    sampling_rate_new: the new sampling rate of the returned data
    
    """
    #decimation factor
    r = int(sampling_rate/sampling_rate_new)    
        
        
    #create array for downsampled data: data_downsampled
    #data_downsampled = np.zeros(np.shape(data_lfp[::r,:,:]))
    
        
    #slice the data array for obtaining every r-th sample:
    data_downsampled = data[::r,:,:]
    
    return data_downsampled
    


def offset_removal (data, sampling_rate_new = 500, prestim_offset_x1 = -1, prestim_offset_x2 = 0):
    """returns data, subtracts the mean of a baseline period from all sample points of a trial
    
    Parameters
    ----------
    data: input array, numpy array (samples x channels x trials)
    
    prestim_offset_x1: in s, start for time range used to compute the offset for each trial rel. to stim artefact
    
    prestim_offset_x2: in s, stop for time range used to compute the offset for each trial rel. to stim artefact
    
    """
    
    #number of samples from start to stimulation artefact: prestim
    prestim = (np.size(data, axis =0)/2)/sampling_rate_new
    
    #start and stop samples of baseline (before stimulation) for calculating offset: start, stop
    start = int(sampling_rate_new * prestim_offset_x1)+ int(sampling_rate_new * prestim)
    stop =  int(sampling_rate_new * prestim_offset_x2)+ int(sampling_rate_new * prestim)
        
    #Offset computation:
    offset_matrix = data[start:stop,:,:]
    offset = np.mean(offset_matrix, axis = 0)
    
    #create empty array for offset subtracted data: data_subtracted
    data_subtracted = np.zeros(np.shape(data))
    
    #subtract baseline offset from all samples x channels x trials with for loop
    for ch in range(np.size(data, axis = 1)):
        for sw in range (np.size(data, axis = 2)):
            data_subtracted[:,ch,sw] = data[:, ch, sw] - offset[ch, sw]
    
    return data_subtracted
            
        
        
      
 

def trial_rejection(data, sampling_rate_new = 500, prestim_rms_x1 = -1, prestim_rms_x2 = 0, dependent_rms_x1 = -1, dependent_rms_x2 = 0, factor_std =3):
    """returns data (samples x channels x trials) where trials with high threshold voltage artefacts are removed.
    Threshold for rejection is calculated by the averaged root mean square (RMS) of baseline. 
    -->e.g. from -1 to 0 s across trials + 3 standard deviations. 
    returns also array a2 (trials): if value is 1 the trial is conserved/not rejcted for all channels
    
    Parameters
    ----------
    data: input array, numpy array (samples x channels x trials)
    
    prestim_rms_x1: in s, start for time range of baseline used to compute the rms for each trial rel. to stim artefact
    
    prestim_rms_x2: in s, stop for time range of baseline used to compute the rms for each trial rel. to stim artefact    
        
    dependent_rms_x1: in s, start time range of baseline used to compare [s]
    
    dependent_rms_x2: in s, stop time range of baseline used to compare [s]
    
    factor_std: (int), scale factor multiplies with std of rms to calculate threshold for trial rejection. -->high factor rejects less trials
    """
    
        
    
    #number of samples from start to stimulation artefact: prestim
    prestim = (np.size(data, axis =0)/2)/sampling_rate_new
    
    #start and stop samples of baseline (before stimulation) for calculating rms: start1, stop1
    start1 = int(sampling_rate_new * prestim_rms_x1)+ int(sampling_rate_new * prestim)
    stop1  = int(sampling_rate_new * prestim_rms_x2)+ int(sampling_rate_new * prestim)
        
    #start and stop samples of baseline (before stimulation) for comparison with rms: start2, stop2
    start2 = int(sampling_rate_new * dependent_rms_x1)+ int(sampling_rate_new * prestim)
    stop2  = int(sampling_rate_new * dependent_rms_x2)+ int(sampling_rate_new * prestim)

    
    
    ## select the range to calculate the root mean square of baseline: prestim_matrix
    prestim_matrix = data[start1:stop1,:,:]     
          
    ## select the range for comparison with rms: dependent_matrix
    dependent_matrix = data[start2:stop2,:,:]     

    
    ##calculate root-mean-squared of prestim_matrix, dependent_matrix : rms_pre, rms_pre_dependent (array shape: channels x trials)
    rms_pre           = np.sqrt(np.mean(prestim_matrix**2, axis = 0))
    
    rms_pre_dependent = np.sqrt(np.mean(dependent_matrix**2, axis = 0))    
    
            
    
    
    ## DEFINES THRESHOLD FOR REJECTION
        
    ## Mean and standard deviation of RMS from all baselines from all trials : mean_rms_pre, std_rms_pre (array shape: channels)
    
    mean_rms_pre = np.mean(rms_pre, axis =1)
    
    std_rms_pre = np.std(rms_pre,1)
    
    
    # threshold at 3 standard deviations : threshold (array shape: channels)
    threshold = mean_rms_pre + (std_rms_pre * factor_std)
    
    
    
   ##### PERFORMS REJECTIONS AND CONSERVES TRIALS WITH STABLE BASELINE  
   ###Identifies only those trials having baseline RMS below threshold in all channels 
   
    #create empty array of shape rms_dependent for appending channels, trials : a1
    a1 = np.zeros(np.shape(rms_pre_dependent))
   
      
   #loop over channels and trials:
    for ch in range (np.size(rms_pre_dependent, axis =0)):
        for swe in range (np.size(rms_pre_dependent, axis = 1)):
            #rms values samller than threshold are appended as True to a1 array (larger values as False)
            a1[ch, swe] = rms_pre_dependent[ch, swe]     <   threshold[ch] 
            #conserve all those trials which are True (mean = 1) across all channels: a2 (trials)
            a2 = np.mean(a1, axis =0) ==1
            #find trial indices where a2==1: trials_find
            trials_find = np.where(a2 ==1)[0]
    
        
    
   #Extract only trials with baseline RMS below threshold in all channels 
   # in electrophysiological signal with common average referencing 
    eeg_good = data[:,:,trials_find]
    return eeg_good, trials_find


def trials_retained (data, *args):
    """returns array with the conserved trials in respect to trials in input data array (samples x channels x trials)
    
    Parameters
    ----------
    data : input array, numpy array (samples x channels x trials)
    *args  : input array(s), returned by trial_rejection function as trials_find (the conserved trials across channels); 
    
    """
    ###get number of trials from input data (samples x channels x trials): trials (integer scalar)
    trials = np.size(data, axis =2)
    ###create array with trials: trials_arr
    trials_arr = np.arange(trials, dtype =np.int32)
    
    ###loop over *args and slice trials_arr according to values in *args: trials_arr
    for i in args:
        trials_arr = trials_arr[i]
    return trials_arr
    
                     


def plot_trials_vstack1 (arr1, ch=0, width = 80, height =60, time_start =0.2, time_stop = 0.6, s_rate = None, sweep_start = 0, sweep_stop = 7, scalebar = True):
    """plots trials on top off each other
    
    parameters
    ----------
    
    arr1: input array (3-D, samples x channels x sweeps), the  (preprocessed) raw data
    
    ch: the channel number to display (int)
    
    s_rate: the sampling rate of the mua raw data
    
    time_start: the start time (seconds) of the sweep to plot, before  stimulation
    
    time_stop: the stop time (seconds) of the sweep to plot, after stimulation
    
    sweep_start: the start number of the sweep to plot (int)    

    sweep_stop: the stop number of the sweep to plot (int)

    fnt: fnt size (int) of labels and title 

    title: the figure title (str)
    
    width: the figure width (int)
    
    height: the figure height (int)  
    
    """
    #Paramters, delete after func completed
    #arr1 = mua_raw_wake
    #width = 80
    #height =60
    #time_start =0.2
    #time_stop = 0.6
    #s_rate = fs
    #sweep_start = 0
    #sweep_stop = 7
    
    
    
    #slice raw data:  
    mua_raw_sl = slice_array(arr1, s_rate =fs, pre_stim=time_start, post_stim=time_stop)
    mua_raw_sl = mua_raw_sl[ch]
    
    
    #create time data for x axis: time
    time = np.linspace(-time_start, time_stop, num = int((time_start + time_stop)* s_rate)+1)

    fig,ax= plt.subplots(figsize=(mm_to_inch(width), mm_to_inch(height)))
    
    
    # use a custom colormap to create a different color for every sweep
    cm = plt.get_cmap("winter")
    colors = [cm(x/(sweep_stop-sweep_start)) for x in range(sweep_start, sweep_stop)]
    
    for i,ii in zip (range (sweep_start, sweep_stop), range(sweep_stop-sweep_start)):
    
        x_data = time + 0.00 * ii
        y_data = mua_raw_sl[:,0,i] + 130 * ii
        
        ax.plot(x_data, y_data, color = colors[ii], alpha =0.5)
        
    fontprops = fm.FontProperties(size=18)
    
    if scalebar == True:
        AnchoredScaleBar(ax.transData)
        #scalebar_v = AnchoredSizeBar(ax.transData, 50, '50 uV s', 'lower right', pad=0.1, color='black',frameon=False, size_vertical=1, fontproperties=fontprops)
        
        sb = add_scalebar(ax, matchx=False, matchy = False)
        ax.add_artist(sb)
    
    fig.gca().axis('off')
    fig.tight_layout()
    fig.show()   
    
    
    
    
    

 
def plot_trials_vstack (data, file_name, save_to_path = '_', file_ending = '_', sampling_rate = 20000, x_start = -2, x_end = 2,y_start=-5000, y_end=1000, scalex = 0.2, scaley=200,  channel_idx = 0, channel_name =0, trial_start =0, number_trials = 10, width=None, height =None, scalebar = True, show_fig = True):
    """
    plots a range of trials and its mean; input data (samples x channels x trials) are returned by slice_array_for_conv function.
    
    Parameters
    ----------
    
    data: npy array with data arranged in samples x channels x trials
    
    file_name: the filename of the recorded data. provided by files_to_list function
    
    save_to_path: file_path for saving raw data traces, provide as raw string, e.g. 
    
    new_file_ending: provide string to save as new figure when function is caled again and one of function paramters was changed (e.g. x_start)
    
    sampling_rate: default = 20000
    
    x_start: default = 0     --> start value for range of time to plot
    
    x_end: default = 1       --> end value for range of time to plot
   
    y_start: default = 0     --> start value for range of y signal to plot
    
    y_end: default = 1       --> end value for range of y signal to plot   
    
    scalex: default = 0.2   --> scale bar size x-axis in seconds
    
    scaley= default = 200   --> scale bar size y-axis
    
    channel_idx: default = 0         --> channel to visualize, provide channel index of input array    
      
    channel_name: default = 0         --> channel name in title of figure    
    
    trial_start: default =0       --> start value for range of trials to plot 
        
    number trials: default =10    --> number of trials to plot   
    
    width: the figure width in mm

    height: the figure height in mm    
    """
    
    
    length_trial_s = np.size(data, axis = 0)/sampling_rate
    start = (length_trial_s/2)*-1
    stop = (length_trial_s/2)
   
    
    
    ###create time vector for plotting x-axis: time
    time = np.linspace(start, stop, num = np.size(data, axis =0), endpoint=False )
    
    fig, ax = plt.subplots(figsize = (mm_to_inch(width), mm_to_inch(height)))
    
    cm = plt.get_cmap("winter")
    
    #defines colored traces
    #colors = [cm(x/((trial_start+ number_trials)-trial_start)) for x in range(trial_start, number_trials)]

    #plots a range of trials + mean of all trials 
    for trial, i in zip (range (trial_start, trial_start + number_trials), range((trial_start+ number_trials)-trial_start)):
        y_data = data[ : , channel_idx, trial] + 500 * i
        #if traces need to be colored....
        #ax.plot(time, y_data, color= colors[i], linewidth=0.5)
        ax.plot(time, y_data, 'k', linewidth=0.5)
        
    #mean of al trials: mean_trials
    
    plt.rcParams.update({'font.size': 22})
    plt.xlabel('Time(s)', fontsize=22)
    plt.ylabel('Amplitude (uV)', fontsize=22)
    plt.title ('ERP of ' + str(number_trials) + ' single trials from channel ' + str(channel_name))
    plt.tick_params(labelsize=18)
    #plt.axvline(x = 0, linewidth=1, color='red', linestyle='--')
    plt.axis([x_start, x_end, y_start, y_end])
    
    #add scalebar
    if scalebar == True:
        #AnchoredScaleBar(ax.transData)
        #scalebar_v = AnchoredSizeBar(ax.transData, 50, '50 uV s', 'lower right', pad=0.1, color='black',frameon=False, size_vertical=1, fontproperties=fontprops)
        
        sb = add_scalebar(ax, matchx=False, matchy = False, sizex= scalex, sizey=scaley)
        ax.add_artist(sb)
        ax.
    
    #ax.get_yaxis().set_visible(False)
    
 
   
    if show_fig == True:
        plt.show()
    else:
        plt.savefig(save_to_path + file_name + file_ending + "_ch" +str(channel_name)+"_" + ".png", format="PNG")
        plt.close()
    



def mm_to_inch(value):
    """converts mm to inch (for use in e.g. plt.figure(figsize=(mm_to_inch(20), mm_to_inch(10))"""
    return (value/25.4)

    

class AnchoredScaleBar(AnchoredOffsetbox):
    def __init__(self, transform, sizex=0.2, sizey=200, labelx='', labely='', loc=3,
                 pad=0.1, borderpad=0.1, sep=1, prop=None, barcolor="black", barwidth=None, 
                 **kwargs):
        """
        Draw a horizontal and/or vertical  bar with the size in data coordinate
        of the give axes. A label will be drawn underneath (center-aligned).
        - transform : the coordinate frame (typically axes.transData)
        - sizex,sizey : width of x,y bar, in data units. 0 to omit
        - labelx,labely : labels for x,y bars; None to omit
        - loc : position in containing axes
        - pad, borderpad : padding, in fraction of the legend font size (or prop)
        - sep : separation between labels and bars in points.
        - **kwargs : additional arguments passed to base class constructor
        """
        from matplotlib.patches import Rectangle
        from matplotlib.offsetbox import AuxTransformBox, VPacker, HPacker, TextArea, DrawingArea
        bars = AuxTransformBox(transform)
        if sizex:
            bars.add_artist(Rectangle((0,0), sizex, 0, ec=barcolor, lw=barwidth, fc="none"))
        if sizey:
            bars.add_artist(Rectangle((0,0), 0, sizey, ec=barcolor, lw=barwidth, fc="none"))

        if sizex and labelx:
            self.xlabel = TextArea(labelx, minimumdescent=False)
            bars = VPacker(children=[bars, self.xlabel], align="center", pad=0, sep=sep)
        if sizey and labely:
            self.ylabel = TextArea(labely)
            bars = HPacker(children=[self.ylabel, bars], align="center", pad=0, sep=sep)

        AnchoredOffsetbox.__init__(self, loc, pad=pad, borderpad=borderpad,
                                   child=bars, prop=prop, frameon=False, **kwargs)

        
def add_scalebar(ax, matchx=False, matchy=False, hidex=True, hidey=True, **kwargs):
    """ Add scalebars to axes
    Adds a set of scale bars to *ax*, matching the size to the ticks of the plot
    and optionally hiding the x and y axes
    - ax : the axis to attach ticks to
    - matchx,matchy : if True, set size of scale bars to spacing between ticks
                    if False, size should be set using sizex and sizey params
    - hidex,hidey : if True, hide x-axis and y-axis of parent
    - **kwargs : additional arguments passed to AnchoredScaleBars
    Returns created scalebar object
    """
    def f(axis):
        l = axis.get_majorticklocs()
        return len(l)>1 and (l[1] - l[0])
    
    if matchx:
        kwargs['sizex'] = f(ax.xaxis)
        kwargs['labelx'] = str(kwargs['sizex'])
    if matchy:
        kwargs['sizey'] = f(ax.yaxis)
        kwargs['labely'] = str(kwargs['sizey'])
        
    sb = AnchoredScaleBar(ax.transData, **kwargs)
    ax.add_artist(sb)

    if hidex : ax.xaxis.set_visible(False)
    if hidey : ax.yaxis.set_visible(False)
    if hidex and hidey: ax.set_frame_on(False)

    return sb

    
    


 
def plot_data (data, file_name, save_to_path = '_', file_ending = '_', sampling_rate = 20000, x_start = -2, x_end = 2,y_start=-5000, y_end=1000, channel_idx = 0, channel_name =0, trial_start =0, number_trials = 10, show_fig = True):
    """
    plots a range of trials and its mean; input data (samples x channels x trials) are returned by slice_array_for_conv function.
    
    Parameters
    ----------
    
    data: npy array with data arranged in samples x channels x trials
    
    file_name: the filename of the recorded data. provided by files_to_list function
    
    save_to_path: file_path for saving raw data traces, provide as raw string, e.g. 
    
    new_file_ending: provide string to save as new figure when function is caled again and one of function paramters was changed (e.g. x_start)
    
    sampling_rate: default = 20000
    
    x_start: default = 0     --> start value for range of time to plot
    
    x_end: default = 1       --> end value for range of time to plot
   
    y_start: default = 0     --> start value for range of y signal to plot
    
    y_end: default = 1       --> end value for range of y signal to plot   
    
    channel_idx: default = 0         --> channel to visualize, provide channel index of input array    
      
    channel_name: default = 0         --> channel name in title of figure    
    
    trial_start: default =0       --> start value for range of trials to plot 
        
    number trials: default =10    --> number of trials to plot   
        
    """
    
    
    length_trial_s = np.size(data, axis = 0)/sampling_rate
    start = (length_trial_s/2)*-1
    stop = (length_trial_s/2)
   
    
    
    ###create time vector for plotting x-axis: time
    time = np.linspace(start, stop, num = np.size(data, axis =0), endpoint=False )
    
    #plots a range of trials + mean of all trials 
    for trial in range (trial_start, trial_start + number_trials):
        _ = plt.figure(num =1, figsize = (15,10))
        #plt.plot(time, data[ : , channel_idx, trial], 'tab:gray', linewidth=0.5)
        plt.plot(time, data[ : , channel_idx, trial],  linewidth=0.5)
    #mean of al trials: mean_trials
    mean_trials = np.mean (data[:,:,:], axis = 2)
    
    plt.plot(time, mean_trials[:,channel_idx], 'k', linewidth = 3) 
    plt.rcParams.update({'font.size': 22})
    plt.xlabel('Time(s)', fontsize=22)
    plt.ylabel('Amplitude (uV)', fontsize=22)
    plt.title ('Mean ERP and ' + str(number_trials) + ' single trials from channel ' + str(channel_name))
    plt.tick_params(labelsize=18)
    #plt.axvline(x = 0, linewidth=1, color='red', linestyle='--')
    plt.axis([x_start, x_end, y_start, y_end])
 
    
    if show_fig == True:
        plt.show()
    else:
        plt.savefig(save_to_path + file_name + file_ending + "_ch" +str(channel_name)+"_" + ".png", format="PNG")
        plt.close()
    

```
