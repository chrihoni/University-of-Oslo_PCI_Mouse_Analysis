```python
    
def plot_pci_st(x,y, z, y_err, z_err, y_err_low_eeg,  title = '_', fig_height = 20, fig_width= 10):
    """    

    Parameters
    ----------
    x : (1-D, array or list), the time data returned by sliding_window func
    
    y : (1-D, array or list), the pci_st data returned by sliding_window func for wake state
    
    z : (1-D, array or list), the pci_st data returned by sliding_window func for anesthetized state
    
    y_err : the error (e.g.std) plotted around y
    
    z_err : the error (e.g.std) plotted around z
    
    y_err_low_eeg : the 95% CI for difference in means, lower percentile (2.5%) >0  plotted  for eeg/lfp
    
    plots x (time), y(control in black), z(drug in red) and  labels and title 
    """
    
    #parameters for testing
    #x = time_sl_window
    #y = pci_st_100ms_w
    #title = 'example'
    #convert input lists to arrays
    x_arr, y_arr, z_arr, y_err_arr, z_err_arr, y_err_low_eeg  = np.array(x), np.array(y), np.array(z), np.array(y_err), np.array(z_err), np.array(y_err_low_eeg)
    
    #convert x_arr in ms to s
    x_arr = x_arr/1000
    
    plt.figure(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))

        
        ##calculate lower error: le_y, le_z
    le_y = y_arr - y_err_arr
    le_z = z_arr - z_err_arr

    
    #find indices where values < 0
    ix_y = np.where(le_y < 0)
    ix_z = np.where(le_z < 0)
    
    #set indices (ix) in le to 0
    le_y[ix_y[0]] = 0
    le_z[ix_z[0]] = 0
    
    ##calculate upper error: ue 
    ue_y = y_arr + y_err_arr
    ue_z = z_arr + z_err_arr
    
    

    
    
    _ = plt.plot(x_arr, y_arr, 'r',  label = 'Wake')
    _ = plt.plot(x_arr, z_arr, 'b', label = 'Sevo')
        # plot the shaded range of the error       
    _ = plt.fill_between(x_arr, le_y, ue_y, color='r', alpha=.1)
    _ = plt.fill_between(x_arr, le_z, ue_z, color='b', alpha=.1)
    #plot bar indicating diff in 95% CI of means
    _=  plt.fill_between(x_arr, 10, 10.1, where= (y_err_low_eeg >0) , facecolor='k', label='Δ 95% CI > 0')


    
    plt.xlabel('Time (s) ', fontsize = 25)
    plt.ylabel('PCI_st (arb. unit)', fontsize=25)    
    plt.title (title, fontsize=20)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.axis([-0.02, 0.55, -1, 12])
    plt.grid()
    plt.legend(fontsize=20, loc="center right", bbox_to_anchor=(0.95, 0.65))
    #plt.legend(fontsize=20)

    plt.tight_layout()
    plt.show()
    


def CI_bigger_0 (y_err_low_eeg, y_err_low_lfp, time_resp):
    """returns timepoints (1D array) when  95% CI  for difference in means (wake_sevo)> 0 in PCIst EEG and LFP   """
    eeg_low_bigger_0 = np.where(np.array(y_err_low_eeg) >0)[0]
    eeg_time_bigger_0 = np.array(time_resp)[eeg_low_bigger_0]
    
    lfp_low_bigger_0 = np.where(np.array(y_err_low_lfp) >0)[0]
    lfp_time_bigger_0 = np.array(time_resp)[lfp_low_bigger_0]
    
    
    eeg_lfp_bigger_0 = [np.where(lfp_time_bigger_0 ==i )[0][0] for i in eeg_time_bigger_0 if np.size(np.where(lfp_time_bigger_0 ==i )[0])==1 ]
    
    return lfp_time_bigger_0[eeg_lfp_bigger_0]

   
def plot_pci_st_diff(x, y_err_low_eeg, y_err_high_eeg, y_err_low_lfp, y_err_high_lfp, time_overlap, title = '_', fig_height = 20, fig_width= 10):
    """plots confidence  interval over time specified by  y_err_lowand y_err_high  

    Parameters
    ----------
    x : (1-D, array or list), the time data returned by sliding_window func
    
    y_err_low_eeg : the lower error  plotted  for eeg
    
    y_err_high_eeg: the upper error  plotted for eeg
    
    
    y_err_low_lfp : the lower error  plotted  for lfp
    
    y_err_high_lfp: the upper error  plotted for lfp
    
    time_overlap: the time when 95% CI for difference in means are > 0 for both EEG and LFP

    """
    
    #parameters for testing
    #x = time_sl_window
    #y = pci_st_100ms_w
    #title = 'example'
    #convert input lists to arrays
    x_arr, y_err_l_eeg,  y_err_h_eeg, y_err_l_lfp,  y_err_h_lfp  = np.array(x), np.array(y_err_low_eeg),  np.array(y_err_high_eeg), np.array(y_err_low_lfp),  np.array(y_err_high_lfp)
    
    #convert x_arr in ms to s
    x_arr = x_arr/1000
    
    plt.figure(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))

        
    
        # plot the shaded range of the error       
    _ = plt.fill_between(x_arr, y_err_l_eeg, y_err_h_eeg, alpha=.6, label='Δ 95% CI EEG', color='green')
    _ = plt.fill_between(x_arr, y_err_l_lfp, y_err_h_lfp, alpha=.6, label='Δ 95% CI LFP', color='orange')
    _=  plt.fill_between(x_arr, 8, 8.1, where= (y_err_l_eeg >0) & (y_err_l_lfp >0) , facecolor='k', label='Δ 95% CI of\nEEG & LFP > 0')
    
    plt.xlabel('Time (s) ', fontsize = 25)
    plt.ylabel('Diff. in PCI_st (arb. unit)', fontsize=25)    
    plt.title (title, fontsize=20)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.axis([-0.02, 0.55, -2, 10])
    plt.grid()
    plt.legend(fontsize=20, bbox_to_anchor=(0.95, 0.8))
    plt.tight_layout()
    plt.show()
    




def plot_paired_dot (before, after, title='PCIst_80to600ms_eeg', y_label='PCIst', y_axis_start= 0, y_axis_stop=0.3, fig_height = 20, fig_width= 10):
    """returns plot with paired data before and after treatment + box plot on top, and p-value of permutation hypothesis test of paired differences
    
    Parameters
    ----------
    before: Pandas dataframe (1-D) or array-like containing data of before treatment group
    
    after:  Pandas dataframe (1-D) or array-like containing data of after treatment group
    
    """
    
    #check if object is a pd.series, if yes append values to list
    
    if isinstance(before, pd.Series) == True:
    #convert dataframes to arrays
        before = before.values.tolist()
        after =  after.values.tolist()
    
    
    #calculates p_value
    p = hypothesis_permutation_test_paired_diff(before, after)
    
    fig, ax = plt.subplots(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))

    
    #plotting the points before treatment
    _= plt.scatter(np.zeros(len(before)), before, alpha=0.4,s = 90, c='red')
    #plotting the points after treatment 
    _= plt.scatter(np.ones(len(after)), after, alpha=0.4, s = 90, c='blue')
    

    ####plot lines connecting each paired before and after treatment
    #loop over all indices of input array
    for i in range(len(before)):
        _ = ax.plot([0,1], [before[i], after[i]], c ='k', linewidth=0.2)
    
    #plots p-value
    _ = ax.text(0.45,0.95, 'p = '+str(p), dict(size=15), transform=ax.transAxes)
   
    #plot a boxplot on top of data
    _ = ax.boxplot([before, after], positions = [0,1])
    
    #rename the xlabels
    plt.xticks([0,1], ['Wake', 'Sevo'], fontsize=20)
    plt.yticks(fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.title(str(title), fontsize=20)
    plt.axis([-0.2,1.2,y_axis_start, y_axis_stop])
    plt.tight_layout()
    
    plt.gca().spines['top'].set_visible(True)
    plt.gca().spines['right'].set_visible(True)
    #plt.legend()
    plt.show()
    


def mm_to_inch(value):
    """converts mm to inch (for use in e.g. plt.figure(figsize=(mm_to_inch(20), mm_to_inch(10))"""
    return (value/25.4)


def fit_lin_reg(x, y):

    """ fits linear Regression to x,y using scikit learn and returns slope, intercept and R_squared """
    #get values from pandas series and reshape dependent variable
    x1 = x.reshape(-1,1)
    #y =pci_80to600_wake_eeg_nona.values
    
    #fits Linear regression 
    model = LinearRegression().fit(x1,y)
    #extracts r_squared
    r_sq= model.score(x1,y)
    #extracts intercept
    interc= model.intercept_
    #extracts slope
    slp = model.coef_

    #Alternative calculate slope, intercept and r_squared with ols statsmodels and get more detailed stats
    #x = sm.add_constant(x)
    #model1 = sm.OLS(y, x)
    #results= model1.fit()
    #print(results.summary())
    #results.params[0]
    #results.score[]
    #results.rsquared

    return slp, interc, r_sq


def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]


def plot_scatter_linreg_pearson(x, x1, y, y1, linregx1=0,linregx2=30, title=None,fig_width=None, fig_height=None ):
    """scatterplot, linear regression and pearson correlation coef of input arrays """
    #concatenate  x,x1
    x_con = np.concatenate((x,x1))
    #concatenate  y,y1
    y_con = np.concatenate((y,y1))
    
    slope, intercept, r_sq = fit_lin_reg(x_con, y_con)
    
    x3 = np.array([linregx1,linregx2])
    y3 = slope * x3 + intercept
    
    #compute the pearson correlation coefficient
    p = pearson_r(x_con, y_con)
    
    plt.figure(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))

    _ = plt.plot(x, y, marker='.', linestyle='none', color='blue', markersize =20, label='wake')
    _ = plt.plot(x1, y1, marker='.', linestyle='none', color='red', markersize =20, label ='Sevofluran')
    _ = plt.plot(x3, y3, 'k', label ='R_squ: '+ str(round(r_sq, ndigits=2)) + ' \npearson_coeff: ' + str(round(p ,ndigits=2)))
    
    plt.xticks( fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel('PCI_st_EEG', fontsize=20)
    plt.ylabel('PCI_st_LFP', fontsize=20)
    plt.title(title, fontsize=20)
    plt.legend(loc='lower right')
    plt.grid()
    plt.tight_layout()
    plt.show()


def plot_scatter_linreg_slope_one_state(x, y, linregx1=0,linregx2=30, reps=100, label_='_', title=None,fig_width=None, fig_height=None ):
    """scatterplot, linear regression and  hypothesis test on slope of input arrays
    
    parameters
    ----------
    x: input array eeg wake data
    
    y: input array lfp wake data
    
    linregx1: start value of linear model 
    
    linregx2: stop value of linear model
    
    reps: number of permutation replicates calculated
    
    label_: label of data points x/y, e.g. wake/anesthetized
    
    title: ...of figure
    
    fig_width: figure width
    
    fig_height: figure height
    
    """
    
    ##########linear regression###########
    slope, intercept, r_sq = fit_lin_reg(x, y)
    
    x3 = np.array([linregx1,linregx2])
    y3 = slope * x3 + intercept
    
    
    p_val, slope_obs = hypothesis_linreg_slope(x, y, size=reps)    
    print('slope observed:',   np.round(slope, 2))
    print('r_squ :',   np.round(r_sq, 2))
    print('p-value of slope:', np.round(p_val, 2))




    ##########plot data ##########
    plt.figure(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))
    
    _ = plt.plot(x, y, marker='.', linestyle='none', color='green', markersize =20, label=label_)
    _ = plt.plot(x3, y3, 'k', label='R-squ = ' + str(round(r_sq, ndigits=2)) + '\np(slope) = ' + str(round(p_val, ndigits=3)))
    #_ = plt.plot(x3, y3, 'k', label ='R_squ: '+ str(round(r_sq, ndigits=2)) + ' \npearson_coeff: ' + str(round(rho_obs ,ndigits=2))+ '\np = ' + str(round(p_val, ndigits=5)))
    #_ = plt.fill_between(x3, slope_lower, slope_upper, alpha=.1, color='gray')
    
    plt.axis([700,1200,4, 16])
    plt.xticks( fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel('Stimulation electrode depth (um)', fontsize=20)
    plt.ylabel('PCI_st_EEG', fontsize=20)
    plt.title(title, fontsize=20)
    plt.legend(loc='lower right', fontsize=18)
    plt.grid()
    plt.tight_layout()
    plt.show()
    



def plot_scatter_linreg_slope_two_states(x_1, y_1, x_2, y_2, linregx1=0,linregx2=30, reps=100, label_1='_', label_2='_', x_label='_', y_label='_', axis = [500,1200, -1, 25], title=None,fig_width=None, fig_height=None ):
    """scatterplot, linear regression and  hypothesis test on slope of input arrays 
    
    parameters
    ----------
    x_1: input array, 1st state (e.g. wake)
    
    y_1: input array, 1st state (e.g. wake)
    
    x_2: input array, 2nd state (e.g. sevo)
    
    y_2: input array, 2nd state (e.g. sevo)
    
    
    linregx1: start value of linear model 
    
    linregx2: stop value of linear model
    
    reps: number of permutation replicates calculated
    
    label_1:  label of data points x/y, 1st state (e.g. wake)
    
    label_2: label of data points x/y, 2nd state (e.g. sevo)
    
    title: ...of figure
    
    fig_width: figure width
    
    fig_height: figure height
    
    """
    
    ##########linear regression###########
    #1st state 
    slope_1, intercept_1, r_sq_1 = fit_lin_reg(x_1, y_1)
    x3 = np.array([linregx1,linregx2])
    y3 = slope_1 * x3 + intercept_1
    
    #2nd state 
    slope_2, intercept_2, r_sq_2 = fit_lin_reg(x_2, y_2)
    x4 = np.array([linregx1,linregx2])
    y4 = slope_2 * x4 + intercept_2
   
    
    
    ##########hypothesis test, slope###########
    #1st state
    p_val_1, slope_obs_1 = hypothesis_linreg_slope(x_1, y_1, size=reps)    
    print('slope_1 observed:',   np.round(slope_1, 2))
    print('r_squ_1 :',           np.round(r_sq_1, 2))
    print('p-value_1 of slope:', np.round(p_val_1, 2))

    #2nd state
    p_val_2, slope_obs_2 = hypothesis_linreg_slope(x_2, y_2, size=reps)    
    print('slope_2 observed:',   np.round(slope_2, 2))
    print('r_squ_2 :',           np.round(r_sq_2, 2))
    print('p-value_2 of slope:', np.round(p_val_2, 2))



    ##########plot data ##########
    plt.figure(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))
    
    #scatter plot 1st state
    _ = plt.plot(x_1, y_1, marker='.', linestyle='none', color='red', markersize =20,  label=label_1)
    
    #linear regression 1st state
    _ = plt.plot(x3, y3, 'red', label='R^2 = ' + str(round(r_sq_1, ndigits=2)) + '\np(slope) = ' + str(round(p_val_1, ndigits=2)))

    #scatter plot 2nd state
    _ = plt.plot(x_2, y_2, marker='.', linestyle='none', color='blue', markersize =20, label=label_2)
    
    #linear regression 2nd state 
    _ = plt.plot(x4, y4, 'blue', label='R^2 = ' + str(round(r_sq_2, ndigits=2)) + '\np(slope) = ' + str(round(p_val_2, ndigits=2)))

    #_ = plt.plot(x3, y3, 'k', label ='R_squ: '+ str(round(r_sq, ndigits=2)) + ' \npearson_coeff: ' + str(round(rho_obs ,ndigits=2))+ '\np = ' + str(round(p_val, ndigits=5)))
    #_ = plt.fill_between(x3, slope_lower, slope_upper, alpha=.1, color='gray')
    
    plt.axis(axis)
    plt.xticks( fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel(x_label, fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.title(title, fontsize=20)
    plt.legend(loc='upper left', fontsize=15)
    plt.grid()
    plt.tight_layout()
    plt.show()
    





def hypothesis_linreg_slope (x, y, size=1):
    """
    Parameters
    ----------
    x : 1D arr, the x-data
    
    y : 1D arr. the y-data
        
    size : the number of permutation replicates performed. The default is 1.

    Returns
    -------
    p_val : p value of permutation hypothesis test, computes observed slope of linear regression, 
    simulates null hypothesis by calculating perm replicates of slope, 
    and calculates fraction of perm_repl as extreme as the observed slope

    """

    
    #compute the observed slope,  coefficient 1 st order: slope_obs
    slope_obs, intercept, r_sq = fit_lin_reg(x, y)
    #test hypothesis that PCI st is correlated between EEG and LFP data in wake and anethesized states
    # Initialize permutation reps
    perm_reps_slope = np.empty(size)
    # Make permutation reps
    for i in range(size):
        # Scramble the x array
        x_shuff = np.random.permutation(x)
        
        # Compute the slope from permutation replicate
        perm_reps_slope[i], _,_ = fit_lin_reg(x_shuff, y)
    
    if slope_obs > 0:    
    # Compute and print p-value
        p_val = np.sum(perm_reps_slope >= slope_obs) / size
    else:
        p_val = np.sum(perm_reps_slope <= slope_obs) / size
    return p_val, slope_obs




def  polynomial_coeff_sm(x, y):
    """
    
    returns 
    
    - coefficients of 1st order polynomial 
    
    - R2, the R-squared value 
    
    - R2_adj, the adjusted R-square value
    
    -f_pvalue, the p-value of the test statistic, low means linear relationship between price and set of predictors variables
    
    parameters
    ----------
    
    x: input array containing x values for linear fit
    
    y: input array containing y values for linear fit

    """
    
    #Although we are using statsmodel for regression, we’ll use sklearn for generating Polynomial features as it provides simple function to generate polynomials (https://ostwalprasad.github.io/machine-learning/Polynomial-Regression-using-statsmodel.html)
    #set polynomial degree, pass to polynomial features
    polynomial_features= PolynomialFeatures(degree=1)
    #reshape x: x_r
    x_r = x.reshape(-1,1)
    #fit features to x_r
    xp = polynomial_features.fit_transform(x_r)
    #build OLS model
    model1= sm1.OLS(y, xp).fit()
    #make prediction, get y-values 
    #y_pred = model1.predict(xp)
    
    #get summary of descriptive stats
    #model1.summary()
    
    ##get list of calculated parameters with dir(model1)
    
    #get model coefficients
    coeff= model1.params
    
    #get R2 and adjusted R2
    R2, R2_adj = model1.rsquared, model1.rsquared_adj
    
    #get f p-value
    f_pvalue = model1.f_pvalue
      
    
    return coeff, R2, R2_adj, f_pvalue


def draw_bs_pairs_diff(x, y, func, size=1, args=()):
    """
    Perform pairs bootstrap for difference of single statistic.

    Parameters
    ----------
    x : array_like
        x-values of data.
    y : array_like
        y-values of data.
    func : function
        Function, with call signature `func(x, y, *args)` to compute
        replicate statistic from pairs bootstrap sample. It must return
        a single, scalar value.
    size : int, default 1
        Number of pairs bootstrap replicates to draw.
    args : tuple, default ()
        Arguments to be passed to `func`.

    Returns
    -------
    output : ndarray
        Bootstrap replicates.
    """
    x, y = _convert_two_data(x, y, min_len=1)

    # Make Numba'd function
    f, numba_success = _make_two_arg_numba_func(func, args)

    if numba_success:
        jit = numba.jit
    else:
        jit = _dummy_jit

    n = len(x)

    @jit(nopython=True)
    def _draw_bs_pairs(x, y):
        # Set up array of indices to sample from
        inds = np.arange(n)

        # Initialize replicates
        bs_replicates = np.empty(size)

        # Generate replicates
        for i in range(size):
            bs_inds = np.random.choice(inds, n)
            bs_x, bs_y = x[bs_inds], y[bs_inds]
            
            #bs_replicates[i] = f(bs_x, bs_y, args)
            
            bs_replicates[i] = func(bs_x) - func(bs_y) 


        return bs_replicates

    return _draw_bs_pairs(x, y)



def _convert_data(data, inf_ok=False, min_len=1):
    """
    Convert inputted 1D data set into NumPy array of floats.
    All nan's are dropped.

    Parameters
    ----------
    data : int, float, or array_like
        Input data, to be converted.
    inf_ok : bool, default False
        If True, np.inf values are allowed in the arrays.
    min_len : int, default 1
        Minimum length of array.

    Returns
    -------
    output : ndarray
        `data` as a one-dimensional NumPy array, dtype float.
    """
    # If it's scalar, convert to array
    if np.isscalar(data):
        data = np.array([data], dtype=np.float)

    # Convert data to NumPy array
    data = np.array(data, dtype=np.float)

    # Make sure it is 1D
    if len(data.shape) != 1:
        raise RuntimeError("Input must be a 1D array or Pandas series.")

    # Remove NaNs
    data = data[~np.isnan(data)]

    # Check for infinite entries
    if not inf_ok and np.isinf(data).any():
        raise RuntimeError("All entries must be finite.")

    # Check to minimal length
    if len(data) < min_len:
        raise RuntimeError("Array must have at least {0:d} non-NaN entries.".format(min_len))

    return data


def _make_two_arg_numba_func(func, func_args):
    """
    Make a Numba'd version of a function that takes two positional
    arguments.

    Parameters
    ----------
    func : function
        Function with call signature `func(x, y, *args)`.
    func_args : tuple
        Tuple of args to use in testing a function call.

    Returns
    -------
    output : Numba'd function (or original function)
        A Numba'd version of the function. If that is not possible,
        returns the original function.
    numba_success : bool
        True if function was successfully jitted.
    """
    try:
        func_numba = numba.jit(func, nopython=True)

        @numba.jit(nopython=True)
        def f(x, args=()):
            return func_numba(x, *args)

        # Attempt function call
        _ = f(np.array([1., 2.]), np.array([1., 2.]), func_args)

        return f, True
    except:
        def f(x, y, args=()):
            return func(x, y, *args)

        return f, False



def _dummy_jit(*args, **kwargs):
    """Dummy wrapper for jitting if numba not applicable."""

    def wrapper(f):
        return f

    def marker(*args, **kwargs):
        return marker

    if (
        len(args) > 0
        and (args[0] is marker or not callable(args[0]))
        or len(kwargs) > 0
    ):
        # @jit(int32(int32, int32)), @jit(signature="void(int32)")
        return wrapper
    elif len(args) == 0:
        # @jit()
        return wrapper
    else:
        # @jit
        return args[0]


def _convert_two_data(x, y, inf_ok=False, min_len=1):
    """
    Converted two inputted 1D data sets into Numpy arrays of floats.
    Indices where one of the two arrays is nan are dropped.

    Parameters
    ----------
    x : array_like
        Input data, to be converted. `x` and `y` must have the same length.
    y : array_like
        Input data, to be converted. `x` and `y` must have the same length.
    inf_ok : bool, default False
        If True, np.inf values are allowed in the arrays.
    min_len : int, default 1
        Minimum length of array.

    Returns
    -------
    x_out : ndarray
        `x` as a one-dimensional NumPy array, dtype float.
    y_out : ndarray
        `y` as a one-dimensional NumPy array, dtype float.
    """
    # Make sure they are array-like
    if np.isscalar(x) or np.isscalar(y):
        raise RuntimeError("Arrays must be 1D arrays of the same length.")

    # Convert to Numpy arrays
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)

    # Check for infinite entries
    if not inf_ok and (np.isinf(x).any() or np.isinf(y).any()):
        raise RuntimeError("All entries in arrays must be finite.")

    # Make sure they are 1D arrays
    if len(x.shape) != 1 or len(y.shape) != 1:
        raise RuntimeError("Input must be a 1D array or Pandas series.")

    # Must be the same length
    if len(x) != len(y):
        raise RuntimeError("Arrays must be 1D arrays of the same length.")

    # Clean out nans
    inds = ~np.logical_or(np.isnan(x), np.isnan(y))
    x = x[inds]
    y = y[inds]

    # Check to minimal length
    if len(x) < min_len:
        raise RuntimeError(
            "Arrays must have at least {0:d} mutual non-NaN entries.".format(min_len)
        )

    return x, y


def swap_random(a, b):
    """
    Randomly swap entries in two arrays.

    Parameters
    ----------
    a : array_like
        1D array of entries to be swapped.
    b : array_like
        1D array of entries to be swapped. Must have the same lengths
        as `a`.

    Returns
    -------
    a_out : ndarray, dtype float
        Array with random entries swapped.
    b_out : ndarray, dtype float
        Array with random entries swapped.
    """
    a, b = _convert_two_data(a, b)

    return _swap_random(a, b)


@numba.jit(nopython=True)
def _swap_random(a, b):
    """
    Randomly swap entries in two arrays.

    Parameters
    ----------
    a : array_like
        1D array of entries to be swapped.
    b : array_like
        1D array of entries to be swapped. Must have the same lengths
        as `a`.

    Returns
    -------
    a_out : ndarray, dtype float
        Array with random entries swapped.
    b_out : ndarray, dtype float
        Array with random entries swapped.
    """
    # Indices to swap
    swap_inds = np.where(np.random.random(size=len(a)) < 0.5)

    # Make copies of arrays a and b for output
    a_out = np.copy(a)
    b_out = np.copy(b)

    # Swap values
    a_out[swap_inds] = b[swap_inds]
    b_out[swap_inds] = a[swap_inds]

    return a_out, b_out

def hypothesis_permutation_test_paired_diff(arr1, arr2, size=10000):
    """returns p-value of permutation hypothesis test using mean difference of paired observations in arr1 and arr2 as test stattistic.
    Null hypothesis: there is no mean difference of paired obeservations between arr1, and arr2 (wake and sevo condition in each animal)
    
    parameters
    ----------
    arr1: (1-D arr or pandas series) condition1 (e.g. wake) in each experiment/animal

    arr2: (1-D arr or pandas series) condition2 (e.g. anesthetized) in each experiment/animal
    
    size: the number of replicates to be drawn
    
    
    
    """
    ###test statistic for hypothesis test --> mean fractional difference between PCI_st wake and Sevo condition: f_mean
    #convert pandas series to array
    a,b = _convert_two_data(arr1, arr2)
    #compute test statistic of observed fractional difference
    f_mean = np.mean(b-a)
    
    #set up array for appending permutation replicates
    perm_rep = np.empty(size)
    #create permutation replicates
    for i in range(size):
        #create permutation sample by swapping entries
        wake_perm, sevo_perm =swap_random(a, b)
        #calculate perm replicate of test statistic
        f_perm = (sevo_perm-wake_perm)
        #store mean of f_perm in perm_rep
        perm_rep[i] = np.mean(f_perm)
        

    #compute p_value --> The test statistic under the null hypothesis is considered to be at least as extreme as what was observed if it is greater than or equal to f_mean
    p = np.sum(perm_rep <= f_mean)/size
    return p

```
