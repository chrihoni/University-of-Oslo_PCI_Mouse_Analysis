# 2) Functions

```python


def files_to_list (directory = r"D:\Data_scripts\MUA_doubleshank_Neuronexus_probes\2020-12-07", filename_ending = "Rec_7_201207_132859.rhs" ):
    """returns absolute file_paths of data files as list of strings
    
    Parameters
    ----------
    directory: specify the directory where files are located; must be string or converted to raw string (r'...')
    
    filename_ending:  specifiy the file ending e.g. '.npy' , 'NOT_rm.npy'
    """
    
    ####change to directory where files are located, if system can not find the path try to convert to raw string, r'....'
    os.chdir(directory)
    ####empty list for filepaths: file_path
    file_path = []
    file_name = []
    ####for loop appends filename with specific ending to file_path
    for filename in os.listdir(directory):
       if filename.endswith(filename_ending):
           ###creates absolute filepath by merging directory with filename
           file_path.append(os.path.join(directory, filename))
           file_name.append(filename)
    return file_path, file_name



def plot_trials_vstack (data, file_name, save_to_path = '_', file_ending = '_', title='_', sampling_rate = 20000, x_start = -2, x_end = 2,y_start=-5000, y_end=1000,spacing= 500,x_bar=0.2, y_bar= 200, channel_idx = 0, channel_name =0, trial_start =0, number_trials = 10, width=None, height =None, scalebar = True, show_fig = True):
    """
    plots a range of trials and its mean; input data (samples x channels x trials) are returned by slice_array_for_conv function.
    
    Parameters
    ----------
    
    data: npy array with data arranged in samples x channels x trials
    
    file_name: the filename of the recorded data. provided by files_to_list function
    
    save_to_path: file_path for saving raw data traces, provide as raw string, e.g. 
    
    new_file_ending: provide string to save as new figure when function is caled again and one of function paramters was changed (e.g. x_start)
    
    sampling_rate: default = 20000
    
    x_start: default = 0     --> start value for range of time to plot
    
    x_end: default = 1       --> end value for range of time to plot
   
    y_start: default = 0     --> start value for range of y signal to plot
    
    y_end: default = 1       --> end value for range of y signal to plot   
    
    channel_idx: default = 0         --> channel to visualize, provide channel index of input array    
      
    channel_name: default = 0         --> channel name in title of figure    
    
    trial_start: default =0       --> start value for range of trials to plot 
        
    number trials: default =10    --> number of trials to plot   
    
    width: the figure width in mm

    height: the figure height in mm    
    """
    
    
    length_trial_s = np.size(data, axis = 0)/sampling_rate
    start = (length_trial_s/2)*-1
    stop = (length_trial_s/2)
   
    
    
    ###create time vector for plotting x-axis: time
    time = np.linspace(start, stop, num = np.size(data, axis =0), endpoint=False )
    
    fig, ax = plt.subplots(figsize = (mm_to_inch(width), mm_to_inch(height)))
    
    cm = plt.get_cmap("winter")
    
    #defines colored traces
    colors = [cm(x/((trial_start+ number_trials)-trial_start)) for x in range(trial_start, number_trials)]
    #plots a range of trials + mean of all trials 
    for trial, i in zip (range (trial_start, trial_start + number_trials), range((trial_start+ number_trials)-trial_start)):
        y_data = data[ : , channel_idx, trial] + spacing * i
        #if traces need to be colored....
        ax.plot(time, y_data, color= colors[i], linewidth=3)
        #ax.plot(time, y_data, 'k', linewidth=3)

    #mean of al trials: mean_trials
    plt.rcParams.update({'font.size': 22})
    plt.xlabel('Time(s)', fontsize=22)
    plt.ylabel('Amplitude (uV)', fontsize=22)
    plt.title ( title )
    plt.tick_params(labelsize=18)
    #plt.axvline(x = 0, linewidth=1, color='red', linestyle='--')
    plt.axis([x_start, x_end, y_start, y_end])
    
    #add scalebar
    if scalebar == True:
        #AnchoredScaleBar(ax.transData)
        #scalebar_v = AnchoredSizeBar(ax.transData, 50, '50 uV s', 'lower right', pad=0.1, color='black',frameon=False, size_vertical=1, fontproperties=fontprops)
        
        sb = add_scalebar(ax, matchx=False, matchy = False, sizex=x_bar, sizey=y_bar)
        ax.add_artist(sb)
    
    #ax.get_yaxis().set_visible(False)
    
 
    if show_fig == True:
        plt.show()
    else:
        plt.savefig(save_to_path + '/'+ title +  ".svg", format="svg")
        plt.close()
    




def mm_to_inch(value):
    """converts mm to inch (for use in e.g. plt.figure(figsize=(mm_to_inch(20), mm_to_inch(10))"""
    return (value/25.4)

    

class AnchoredScaleBar(AnchoredOffsetbox):
    def __init__(self, transform, sizex=0.2, sizey=200, labelx='', labely='', loc=3,
                 pad=0.1, borderpad=0.1, sep=1, prop=None, barcolor="black", barwidth=None, 
                 **kwargs):
        """
        Draw a horizontal and/or vertical  bar with the size in data coordinate
        of the give axes. A label will be drawn underneath (center-aligned).
        - transform : the coordinate frame (typically axes.transData)
        - sizex,sizey : width of x,y bar, in data units. 0 to omit
        - labelx,labely : labels for x,y bars; None to omit
        - loc : position in containing axes
        - pad, borderpad : padding, in fraction of the legend font size (or prop)
        - sep : separation between labels and bars in points.
        - **kwargs : additional arguments passed to base class constructor
        """
        from matplotlib.patches import Rectangle
        from matplotlib.offsetbox import AuxTransformBox, VPacker, HPacker, TextArea, DrawingArea
        bars = AuxTransformBox(transform)
        if sizex:
            bars.add_artist(Rectangle((0,0), sizex, 0, ec=barcolor, lw=barwidth, fc="none"))
        if sizey:
            bars.add_artist(Rectangle((0,0), 0, sizey, ec=barcolor, lw=barwidth, fc="none"))

        if sizex and labelx:
            self.xlabel = TextArea(labelx, minimumdescent=False)
            bars = VPacker(children=[bars, self.xlabel], align="center", pad=0, sep=sep)
        if sizey and labely:
            self.ylabel = TextArea(labely)
            bars = HPacker(children=[self.ylabel, bars], align="center", pad=0, sep=sep)

        AnchoredOffsetbox.__init__(self, loc, pad=pad, borderpad=borderpad,
                                   child=bars, prop=prop, frameon=False, **kwargs)

        
def add_scalebar(ax, matchx=False, matchy=False, hidex=True, hidey=True, **kwargs):
    """ Add scalebars to axes
    Adds a set of scale bars to *ax*, matching the size to the ticks of the plot
    and optionally hiding the x and y axes
    - ax : the axis to attach ticks to
    - matchx,matchy : if True, set size of scale bars to spacing between ticks
                    if False, size should be set using sizex and sizey params
    - hidex,hidey : if True, hide x-axis and y-axis of parent
    - **kwargs : additional arguments passed to AnchoredScaleBars
    Returns created scalebar object
    """
    def f(axis):
        l = axis.get_majorticklocs()
        return len(l)>1 and (l[1] - l[0])
    
    if matchx:
        kwargs['sizex'] = f(ax.xaxis)
        kwargs['labelx'] = str(kwargs['sizex'])
    if matchy:
        kwargs['sizey'] = f(ax.yaxis)
        kwargs['labely'] = str(kwargs['sizey'])
        
    sb = AnchoredScaleBar(ax.transData, **kwargs)
    ax.add_artist(sb)

    if hidex : ax.xaxis.set_visible(False)
    if hidey : ax.yaxis.set_visible(False)
    if hidex and hidey: ax.set_frame_on(False)

    return sb




```
