```python
def mm_to_inch(value):
    """converts mm to inch (for use in e.g. plt.figure(figsize=(mm_to_inch(20), mm_to_inch(10))"""
    return (value/25.4)


def _dummy_jit(*args, **kwargs):
    """Dummy wrapper for jitting if numba not applicable."""

    def wrapper(f):
        return f

    def marker(*args, **kwargs):
        return marker

    if (
        len(args) > 0
        and (args[0] is marker or not callable(args[0]))
        or len(kwargs) > 0
    ):
        # @jit(int32(int32, int32)), @jit(signature="void(int32)")
        return wrapper
    elif len(args) == 0:
        # @jit()
        return wrapper
    else:
        # @jit
        return args[0]


def _convert_two_data(x, y, inf_ok=False, min_len=1):
    """
    Converted two inputted 1D data sets into Numpy arrays of floats.
    Indices where one of the two arrays is nan are dropped.

    Parameters
    ----------
    x : array_like
        Input data, to be converted. `x` and `y` must have the same length.
    y : array_like
        Input data, to be converted. `x` and `y` must have the same length.
    inf_ok : bool, default False
        If True, np.inf values are allowed in the arrays.
    min_len : int, default 1
        Minimum length of array.

    Returns
    -------
    x_out : ndarray
        `x` as a one-dimensional NumPy array, dtype float.
    y_out : ndarray
        `y` as a one-dimensional NumPy array, dtype float.
    """
    # Make sure they are array-like
    if np.isscalar(x) or np.isscalar(y):
        raise RuntimeError("Arrays must be 1D arrays of the same length.")

    # Convert to Numpy arrays
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)

    # Check for infinite entries
    if not inf_ok and (np.isinf(x).any() or np.isinf(y).any()):
        raise RuntimeError("All entries in arrays must be finite.")

    # Make sure they are 1D arrays
    if len(x.shape) != 1 or len(y.shape) != 1:
        raise RuntimeError("Input must be a 1D array or Pandas series.")

    # Must be the same length
    if len(x) != len(y):
        raise RuntimeError("Arrays must be 1D arrays of the same length.")

    # Clean out nans
    inds = ~np.logical_or(np.isnan(x), np.isnan(y))
    x = x[inds]
    y = y[inds]

    # Check to minimal length
    if len(x) < min_len:
        raise RuntimeError(
            "Arrays must have at least {0:d} mutual non-NaN entries.".format(min_len)
        )

    return x, y


def swap_random(a, b):
    """
    Randomly swap entries in two arrays.

    Parameters
    ----------
    a : array_like
        1D array of entries to be swapped.
    b : array_like
        1D array of entries to be swapped. Must have the same lengths
        as `a`.

    Returns
    -------
    a_out : ndarray, dtype float
        Array with random entries swapped.
    b_out : ndarray, dtype float
        Array with random entries swapped.
    """
    a, b = _convert_two_data(a, b)

    return _swap_random(a, b)


@numba.jit(nopython=True)
def _swap_random(a, b):
    """
    Randomly swap entries in two arrays.

    Parameters
    ----------
    a : array_like
        1D array of entries to be swapped.
    b : array_like
        1D array of entries to be swapped. Must have the same lengths
        as `a`.

    Returns
    -------
    a_out : ndarray, dtype float
        Array with random entries swapped.
    b_out : ndarray, dtype float
        Array with random entries swapped.
    """
    # Indices to swap
    swap_inds = np.where(np.random.random(size=len(a)) < 0.5)

    # Make copies of arrays a and b for output
    a_out = np.copy(a)
    b_out = np.copy(b)

    # Swap values
    a_out[swap_inds] = b[swap_inds]
    b_out[swap_inds] = a[swap_inds]

    return a_out, b_out

def hypothesis_permutation_test_paired_diff(arr1, arr2, size=10000):
    """returns p-value of permutation hypothesis test and mean difference (arr2-arr1) using mean difference of paired observations in arr1 and arr2 as test stattistic.
    Null hypothesis: there is no mean difference of paired obeservations between arr1, and arr2 (wake and sevo condition in each animal)
    
    parameters
    ----------
    arr1: (1-D arr or pandas series) condition1 (e.g. wake) in each experiment/animal

    arr2: (1-D arr or pandas series) condition2 (e.g. anesthetized) in each experiment/animal
    
    size: the number of replicates to be drawn
    
    
    
    """
    ###test statistic for hypothesis test --> mean fractional difference between PCI_st wake and Sevo condition: f_mean
    #convert pandas series to array
    a,b = _convert_two_data(arr1, arr2)
    #compute test statistic of observed fractional difference
    f_mean = np.mean(b-a)
    
    #set up array for appending permutation replicates
    perm_rep = np.empty(size)
    #create permutation replicates
    for i in range(size):
        #create permutation sample by swapping entries
        wake_perm, sevo_perm =swap_random(a, b)
        #calculate perm replicate of test statistic
        f_perm = (sevo_perm-wake_perm)
        #store mean of f_perm in perm_rep
        perm_rep[i] = np.mean(f_perm)
        

    #compute p_value --> The test statistic under the null hypothesis is considered to be at least as extreme as what was observed if it is greater than or equal to f_mean
    p = np.sum(perm_rep <= f_mean)/size
    return p, f_mean


def plot_paired_dot (before, after, title='PCIst_80to600ms_eeg', y_label='PCIst', y_axis_start= 0, y_axis_stop=0.3, fig_height = 20, fig_width= 10):
    """returns plot with paired data before and after treatment + box plot on top, and p-value of permutation hypothesis test of paired differences
    
    Parameters
    ----------
    before: Pandas dataframe (1-D) or array-like containing data of before treatment group
    
    after:  Pandas dataframe (1-D) or array-like containing data of after treatment group
    
    """
    
    #check if object is a pd.series, if yes append values to list
    
    if isinstance(before, pd.Series) == True:
    #convert dataframes to arrays
        before = before.values.tolist()
        after =  after.values.tolist()
    
    
    #calculates p_value
    p = hypothesis_permutation_test_paired_diff(before, after)
    
    fig, ax = plt.subplots(figsize=(mm_to_inch(fig_width), mm_to_inch(fig_height)))

    
    #plotting the points before treatment
    _= plt.scatter(np.zeros(len(before)), before, alpha=0.4,s = 90, c='red')
    #plotting the points after treatment 
    _= plt.scatter(np.ones(len(after)), after, alpha=0.4, s = 90, c='blue')
    

    ####plot lines connecting each paired before and after treatment
    #loop over all indices of input array
    for i in range(len(before)):
        _ = ax.plot([0,1], [before[i], after[i]], c ='k', linewidth=0.2)
    
    #plots p-value
    _ = ax.text(0.45,0.95, 'p = '+str(np.round(p[0],2)), dict(size=15), transform=ax.transAxes)
   
    #plot a boxplot on top of data
    _ = ax.boxplot([before, after], positions = [0,1])
    
    #rename the xlabels
    plt.xticks([0,1], ['Wake', 'Sevo'], fontsize=20)
    plt.yticks(fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.title(str(title), fontsize=20)
    plt.axis([-0.2,1.2,y_axis_start, y_axis_stop])
    plt.tight_layout()
    
    plt.gca().spines['top'].set_visible(True)
    plt.gca().spines['right'].set_visible(True)
    #plt.legend()
    plt.show()
    
    
    

def  polynomial_coeff(df, x, column_select=str, column_label=str ):
    """
    parameters
    ----------
    returns coefficients of 2nd order polynomial fit to median of specified column
    
    df: input dataframe
    
    x: input array containing x values for polynomial fit
    
    column_select: the selected column (str) on which fitting is performed
    
    column_label: the column (str) containing rows labels used for selecting data

    """
    ##loop over cortex_layer and calculate median and mean for each layer --> x -values for curve fitting
    #make lists for appending median and mean values: lst_median
    layers = df[column_label].unique()
    y =  []
    for i in layers:
        #select rows and apply function
        median = df.loc[(df[column_label] == i), column_select].median()
        #append median values of 'diff_itpc_drop_mean' for each layer to lst_median (starting with layer 1 in acsending order according to  layers list)
        y.append(median)
    y = np.array(y)
    
    ## fit cubic (2nd order) polynomial to median itpc values of lst_median: coeff
    #define y-values, i.e. [0,1,2,3,4]
    coeff = np.polynomial.polynomial.polyfit(x,y, 2)
    return coeff, y



def  polynomial_coeff_sm(df1, x, column_select=str, column_label=str ):
    """
    
    returns 
    
    - coefficients of 2nd order polynomial fit to median of specified column using statsmodels
    
    - the observed median y-values for each layer
    
    - R2, the R-squared value 
    
    - R2_adj, the adjusted R-square value
    
    -f_pvalue, the p-value of the test statistic, low means linear relationship between price and set of predictors variables
    
    parameters
    ----------
    df: input dataframe
    
    x: input array containing x values for polynomial fit
    
    column_select: the selected column (str) on which fitting is performed
    
    column_label: the column (str) containing rows labels used for selecting data

    """
    #test arguments
    #df1 = df_diff_itpc_layer
    #x = np.array([0, 1, 2, 3, 4])
    #column_select='diff_itpc_drop_mean'
    #column_label='cortex_layer'
    

    
    ##loop over cortex_layer and calculate median and mean for each layer --> x -values for curve fitting
    #make lists for appending median and mean values: lst_median
    layers = df1[column_label].unique()
    y =  []
    for i in layers:
        #select rows and apply function
        median = df1.loc[(df1[column_label] == i), column_select].median()
        #append median values of 'diff_itpc_drop_mean' for each layer to lst_median (starting with layer 1 in acsending order according to  layers list)
        y.append(median)
    y = np.array(y)
    
    ## fit cubic (2nd order) polynomial to median itpc values of lst_median: coeff
    #define y-values, i.e. [0,1,2,3,4]
    #coeff = np.polynomial.polynomial.polyfit(x,y, 2)
    #Create a polynomial object from coefficients, (before poly1d) https://numpy.org/doc/stable/reference/routines.polynomials.html
    #model = np.polynomial.Polynomial(coeff)
    #pass to statsmodels a generic equation object, i.e. polynomial object p and fit ols: results
    #dic = dict(y=y, x = x)
    #df_dic = pd.DataFrame(dic)
    #results = sm.ols(formula = 'y ~ model(x)', data =dic).fit()
    
    #results.summary()
    
    #Although we are using statsmodel for regression, we’ll use sklearn for generating Polynomial features as it provides simple function to generate polynomials (https://ostwalprasad.github.io/machine-learning/Polynomial-Regression-using-statsmodel.html)
    #set polynomial degree, pass to polynomial features
    polynomial_features= PolynomialFeatures(degree=2)
    #reshape x: x_r
    x_r = x.reshape(-1,1)
    #fit features to x_r
    xp = polynomial_features.fit_transform(x_r)
    #build OLS model
    model1= sm1.OLS(y, xp).fit()
    #make prediction, get y-values 
    #y_pred = model1.predict(xp)
    
    #get summary of descriptive stats
    #model1.summary()
    
    ##get list of calculated parameters with dir(model1)
    
    #get model coefficients
    coeff= model1.params
    
    #get R2 and adjusted R2
    R2, R2_adj = model1.rsquared, model1.rsquared_adj
    
    #get f p-value
    f_pvalue = model1.f_pvalue
      
    
    return coeff, y, R2, R2_adj, f_pvalue


def draw_perm_rep_polynomial_coeff(df, reps, column_select, column_label):
    """
    parameters
    ----------
    
    df: input dataframe
    
    column_select: the column to use for permutation
    
    column_label: the column containing rows labels used for selecting data for permutation, (for loop iterates over this labels)
    
    """
    
    #test arguments
    #reps = 1
    #column_select = 'diff_itpc_drop_mean'
    #column_label = 'animal_id'
    
    
    df1= df.copy()

    #select unique column labels: label
    label = df1[column_label].unique()
    
    #create empty array for appending permutation replicates: perm_reps
    coeff_reps = np.empty(reps)
    #for loop over reps
    for i in range (reps):
        #for loop over label 
        for ii in label:
            #select rows with label: select 
            select = df1.loc[(df1[column_label]==ii), column_select]
            #shuffle selected rows: shuffle
            shuffle= np.random.permutation(select)
            df1.loc[(df1[column_label]==ii), column_select] = shuffle
        
        #calculate the polynomial coefficient across layers for df1: coeff_rep
        coeff_rep = polynomial_coeff(df1, np.array([0, 1, 2, 3, 4]), column_select='diff_itpc_drop_mean', column_label='cortex_layer')
        coeff_reps[i] = coeff_rep[0][2]
    return coeff_reps


```

#### calculate mean itpc value across channels for each animal and condition and append to new dataframe

```python

                             ####plot itpc_drop lfp data across cortical layers - select data in dataframe####
                             
###plot itpc for each layer/condition/animal 

##create dataframe to append median of itpc_drop for each layer: df_itpc_layer
df_itpc_layer = pd.DataFrame(columns=(['itpc_drop_median', 'itpc_drop_mean', 'cortex_layer', 'state', 'animal_id'])) 

## select rows for itpc for each layer/condition/animal 

#get entries of animal_ids: animal_id
animal_id = df_summary['animal_id'].unique()

#create entries of layers: layers
layers = ['1', '2/3', '4', '5', '6']

#create list with entries of state which starts with ['Wake', 'Anes']: state_startswith
state_startswith = ['Wake', 'Anes']



#loop over animal_id, state, layers
for i in animal_id:
    for ii in state_startswith:
        for iii in layers:       
            itpc = df_summary.loc[(df_summary['state_'].str.startswith(ii)) & (df_summary['eeg_or_lfp']=='lfp') & (df_summary['animal_id'] ==i) & (df_summary['cortex_layer'].str.startswith(iii)), 'itpc_drop'] 
            #drop Nan values
            #itpc = itpc.dropna()
            #calculate mean 
            itpc_mean= itpc.mean()
           
            #calculate median 
            itpc_median= itpc.median()
           
            #create dictionary and append values: d
            d = dict(itpc_drop_median = itpc_median, itpc_drop_mean = itpc_mean, cortex_layer = iii, state = ii, animal_id = i)
            
            #make pd.series from d, convert to dictionary and create dataframe: df
            df = pd.DataFrame(dict([(k, pd.Series(v)) for k,v in d.items() ]))
            
            #append df to df_itpc_layer: df_itpc_layer
            df_itpc_layer = df_itpc_layer.append(df, ignore_index=True)


```

#### calculate differeence in itpc drop between wake and anesthesia for each layer and animal  and append to new dataframe

```python
##create dataframe to append median of itpc_drop for each layer: df_itpc_layer
df_diff_itpc_layer = pd.DataFrame(columns=([ 'diff_itpc_drop_mean', 'cortex_layer', 'animal_id'])) 

## select rows for itpc for each layer/condition/animal 

#get entries of animal_ids: animal_id
animal_id = df_itpc_layer['animal_id'].unique()

#create entries of layers: layers
layers = ['1', '2/3', '4', '5', '6']

#create continous entries of layers for curve fitting (layers are then y coordinates): y
y = np.array([0, 1, 2, 3, 4])


#loop over animal_id, layers and subtract itpc_anes from itpc_wake
for i in animal_id:
    for ii in layers:       
       
        #select itpc during wakefulness
        itpc_wake = df_itpc_layer.loc[(df_itpc_layer['state'].str.startswith('Wake')) &  (df_itpc_layer['animal_id'] ==i) & (df_itpc_layer['cortex_layer'].str.startswith(ii)), 'itpc_drop_mean'] 
        
        #select itpc during anesthesia
        itpc_anes = df_itpc_layer.loc[(df_itpc_layer['state'].str.startswith('Anes')) & (df_itpc_layer['animal_id'] ==i) & (df_itpc_layer['cortex_layer'].str.startswith(ii)), 'itpc_drop_mean'] 
        
        
        
        #calculate difference between itpc_wake and itpc_anes: itpc_diff
        itpc_diff = itpc_wake.values - itpc_anes.values
        
        
        #create dictionary and append values: d
        d = dict(diff_itpc_drop_mean = itpc_diff, cortex_layer = ii, animal_id = i)
        
        #make pd.series from d, convert to dictionary and create dataframe: df
        df = pd.DataFrame(dict([(k, pd.Series(v)) for k,v in d.items() ]))
        
        #append df to df_itpc_layer: df_itpc_layer
        df_diff_itpc_layer = df_diff_itpc_layer.append(df, ignore_index=True)
```

#### ####################hypothesis test######################### 

```python
####################hypothesis test#########################

#1)hypothesis: the itpc drop (wake-anesthesia) follows a c-shaped profile with longest itpc in upper and lower cortical layers 
# and shortest itpc in middle layers (layer 4); null hypothesis--> itpc drop is uniformly distributed, no C-shaped distribution
#2) test statistic: fit 2nd order polynomial and extract 2nd order coefficient 
#3) draw permutation replicates of polynom coefficients --> shuffle itpc in each cortical layer/animal then compute median, fit polynomial and store as replicate
#4) calculate probability of finding a test statistic as extreme as the one observed, under the assumption the null hypothesis is true


perm_reps = draw_perm_rep_polynomial_coeff(df_diff_itpc_layer, 1000, column_select='diff_itpc_drop_mean', column_label='animal_id' )

#4) calculate probability of finding a test statistic as extreme as the one observed, under the assumption the null hypothesis is true: p
p = np.sum(perm_reps >= coeff_observed[2])/len(perm_reps)

```
